```python
#drop unnamed column
df=df.drop("Unnamed: 0",axis = 1)
print(df.head())
```

```python
#change weekly study hours column
df["WklyStudyHours"]= df["WklyStudyHours"].str.replace("05-Oct","5 - 10")
df.head()
```

```python
#Gender distribution
plt.figure(figsize=(5,5))
ax = sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plotting the heatmap
plt.figure(figsize=(5, 5))
sns.heatmap(gb, annot=True, vmin=gb.min().min(), vmax=gb.max().max())
print(gb)
plt.title("Heatmap of Scores by Parent's Education Level")
plt.show()

```

```python
# Assuming df is your DataFrame
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()['EthnicGroup']
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()['EthnicGroup']
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()['EthnicGroup']
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()['EthnicGroup']
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()['EthnicGroup']

# Corrected list of counts
mlist = [groupA, groupB, groupC, groupD, groupE]
print(mlist)
# Plotting the pie chart
plt.pie(mlist, labels=['Group A', 'Group B', 'Group C', 'Group D', 'Group E'], autopct='%1.1f%%')
plt.title("Distribution of Ethnic Groups")
plt.show()
```
